```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# get standard deviation of y-axis values
sd = statistics.stdev(y_axis)
sd
```

```python
#make line chart with error bars (sd of y_axis), matlab way
plt.errorbar(x_axis,y_axis,yerr=sd,capsize = 3)
```

```python
#make line chart with error bars (sd of y_axis), obj-or way
fig, ax = plt.subplots()
ax.errorbar(x_axis, y_axis, yerr=sd, capsize=3)
plt.show()
```

```python
#make bar chart with error bars (sd of y_axis), matlab way
plt.bar(x_axis, y_axis, yerr=sd, capsize=3)
```

```python
#make bar chart with error bars (sd of y_axis), obj-or way
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, yerr=sd, capsize=3)
plt.show()
```

```python
#change the major ticks interval on a bar chart, matlab way
plt.barh(x_axis, y_axis)
plt.xticks(np.arange(0, 51, step=5.0)) #toggle comment of this line to see effect
plt.gca().invert_yaxis()
```

```python
#change the major ticks interval on a bar chart, obj-or way
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
ax.invert_yaxis()
# plt.show()
```

```python
#change the minor ticks interval on a bar chart, obj-or way
# x.xaxis.set_minor_locator()

# Increase the size of the plot figure.
fig, ax = plt.subplots(figsize=(8, 8))
ax.barh(x_axis, y_axis)
ax.set_xticks(np.arange(0, 51, step=5.0))
ax.invert_yaxis()

# Create minor ticks at an increment of 1.
ax.xaxis.set_minor_locator(MultipleLocator(1))
# plt.show()
```
